# 1️⃣ Import Libraries

```python
import sys
sys.path.append("../")

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from src.data_loader import load_RAVDNESS_data
from src.feature_extraction import extract_features
```

# 2️⃣ Load Dataset

```python
dataset_path = "../data/raw/RAVDNESS"
df = load_RAVDNESS_data(dataset_path)
print("Dataset Shape:", df.shape)

df.head()
```

# 3️⃣ Extract MFCC Features

```python
X, y = extract_features(df)
print("Feature shape:", X.shape)
print("Labels shape:", y.shape)
```

# 4️⃣ Save Features

```python
np.save("../data/processed/features.npy", X)
np.save("../data/processed/labels.npy", y)
```

# 5️⃣ Feature Heatmap Visualization

```python
plt.figure(figsize=(14,6))
sns.heatmap(X[:100], cmap="coolwarm")
plt.title("MFCC Feature Heatmap")
plt.show()
```

# 6️⃣ Feature Distribution Plot

```python
plt.figure(figsize=(10,5))
sns.histplot(X.flatten(), bins=50, kde=True)
plt.title("MFCC Feature Distribution")
plt.show()
```
