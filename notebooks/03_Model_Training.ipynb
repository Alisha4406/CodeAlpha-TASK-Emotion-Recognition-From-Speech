# 1пёҸвғЈ Import Libraries

```python
import os
import sys
sys.path.append(os.path.abspath(("../")))
```

```python
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from src.model import build_model
```

# 2пёҸвғЈ Load Features

```python
X = np.load("../data/processed/features.npy")
y = np.load("../data/processed/labels.npy")
```

# 3пёҸвғЈ Encode Labels

```python
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
joblib.dump(encoder, "../models/label_encoder.pkl")
```

# 4пёҸвғЈ Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, "../models/scaler.pkl")
```

# 5пёҸвғЈ Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_encoded,
    test_size=0.2,
    random_state=42
)
```

# 6пёҸвғЈ Build Model

```python
model = build_model(
    input_shape=X_train.shape[1],
    num_classes=len(set(y_encoded))
)

model.summary()
```

# 7пёҸвғЈ Train Model

```python
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

# 8пёҸвғЈ Save Model

```python
model.save("../models/emotion_model.h5")
```

```python
import joblib
joblib.dump(history.history, "../models/history.pkl")
```
